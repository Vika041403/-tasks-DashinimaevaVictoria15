```python
cols = [ 'name' , 'gender', 'birth']
years = range(1880, 2011)
pieces = []
for i in years:
    df = pd.read_table('C:\\Users\\Виктория\\Desktop\\soldatov_ks-data-69c734973a6a\\babynames\\yob%d.txt'%i, sep = ',', engine = 'python' , names = cols)
    df['year'] = i
    pieces.append(df)
    data = pd.concat(pieces, ignore_index = True)
data
```
